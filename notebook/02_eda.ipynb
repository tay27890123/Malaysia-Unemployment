```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# ====================== CONFIG ======================
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
pd.set_option('display.float_format', '{:.3f}'.format)

plt.style.use('seaborn-v0_8-whitegrid')

# ====================== PATHS (from ingestion) ======================
ROOT = Path.cwd()
DATA_PROCESSED = ROOT / "data" / "processed"#we locate a folder using path lib 
#then we can directly call this folder and read the data set file
#(this is a folder only not the file)

IMAGE_DIR_02 = ROOT / "image_chart_02_eda"

# Create image folder if missing
IMAGE_DIR_02.mkdir(parents=True, exist_ok=True)
```

```python
# ====================== 1. LOAD CLEAN DATA ======================
df_malaysia = pd.read_csv(DATA_PROCESSED / "malaysia_clean.csv") #here is read the data set(file) use pathlib to call the folder 
#(so here we call that folder and /(filename))

df__malaysia_10years = df_malaysia[df_malaysia["Year"].between(2012,2022)]

print(f" Loaded clean Malaysia data: {df__malaysia_10years.shape[0]} rows x {df__malaysia_10years.shape[1]} columns")
print("\nColumns:\n", df__malaysia_10years.columns.tolist())
df__malaysia_10years.head()
```

```python
print(" Basic Info ")
print()
display(df__malaysia_10years.info())
```

```python
#Summary statistics MALAYSISA UNEMPLOYMENT RATE BY COVID PERIOD
print("\nSummary Statistics by COVID Period ")
summary = df__malaysia_10years.groupby('covid_period').agg({
    'Year': ['min', 'max', 'count'],
    'Unemployment Rate': ['mean', 'min', 'max', 'std'],
    'GDP (billion USD)': ['mean', 'min', 'max'],
    'unemp_change_pct': ['mean', 'max', 'min']
}).round(3)
display(summary)
```

```python
# Summary distrubution of sectors in Malaysia
print("=== Summary Employment rate of Malaysia Sector during covid(2012-2022.10years)===")

#groupby('x') x will become raw like
#x
#1
#2
summary = df__malaysia_10years.groupby('covid_period').agg({
    'Employment Sector: Agriculture': ['mean', 'min', 'max'],
    'Employment Sector: Industry': ['mean', 'min', 'max'],
    'Employment Sector: Services': ['mean', 'min', 'max']
}).round(3)

display(summary)
```

```python
#TIME-SERIES TREND(unemployment rate )(Time Series = lineplot)
fig, ax = plt.subplots(figsize=(12, 7))
sns.lineplot(
    data=df_malaysia, x='Year', y='Unemployment Rate',
    color='red', linewidth=2.5, marker='o', markersize=8, ax=ax)

ax.axvspan(2020, 2022, color='black', alpha=0.3, label='COVID Period (2020–2022)')#shade the vertical area for covid period
ax.axhspan(2, 3.0, color='yellow',alpha=0.2, label="Low Unemployment Rate ")#shade the horizontal area for low unemployment rate 
ax.set_yticks(np.arange(2,5,0.1))
ax.set_title('Malaysia Unemployment Rate: Pre- and Post-COVID (1991–2022)', fontsize=18)
ax.set_xlabel('Year')
ax.set_ylabel('Unemployment Rate (%)')
ax.legend(fontsize=13)
plt.tight_layout()
plt.savefig(IMAGE_DIR_02 / "malaysia_unemployment_trend.png", dpi=300)
plt.show()
```

```python
#TIME-SERIES 10 years (2012-2022)TREND(unemployment rate )(Time Series = lineplot)
fig, ax = plt.subplots(figsize=(12, 7))
sns.lineplot(
    data=df__malaysia_10years, x='Year', y='Unemployment Rate',
    color='blue', linewidth=2.5, marker='o', markersize=8, ax=ax)

ax.axvspan(2020, 2022, color='black', alpha=0.3, label='COVID Period (2020–2022)')#shade the vertical area for covid period
ax.axhspan(2, 3, color='yellow',alpha=0.2, label="Low Unemployment Rate ")#shade the horizontal area for low unemployment rate 
ax.set_yticks(np.arange(2,5,0.1))
ax.set_title('Malaysia Unemployment Rate: Pre- and Post-COVID (2012–2022)', fontsize=18)
ax.set_xlabel('Year')
ax.set_ylabel('Unemployment Rate (%)')
ax.legend(fontsize=13)
plt.tight_layout()
plt.savefig(IMAGE_DIR_02 / "malaysia_unemployment_trend(2012-2022).png", dpi=300)
plt.show()
```

```python
#Distribution plots(Histogram + Boxplot)

fig, ax =plt.subplots(figsize=(12,7))

#Histogram
sns.histplot(data=df__malaysia_10years, x='Unemployment Rate', binwidth=0.1,kde=True,
             hue='covid_period', palette='coolwarm',alpha=0.7,ax=ax)

ax.set_xticks(np.arange(2.4,4.7,0.1))
ax.set_title("Unemployment Rate Distribution Pre-Covid vs Covid", fontsize=20 ,pad=20)
ax.legend(title="Covid period",fontsize=15,labels=["Pre-covid","Covid"])
plt.tight_layout
plt.savefig(IMAGE_DIR_02/"Malaysia_unemployment_histogram_10years.png", dpi=300)
plt.show()

#Boxplot
fig, ax =plt.subplots(figsize=(12,7))
sns.boxplot(data=df__malaysia_10years, x='covid_period', y='Unemployment Rate',
            hue="covid_period", palette=['blue','red'],ax=ax)

ax.set_yticks(np.arange(2.4, 4.8, 0.1))
ax.set_title("Unemployment Rate Distribution: Pre-Covid vs Covid")
ax.set_xlabel("Covid Period (0 = Pre-Covid, 1 = Covid)")
ax.set_ylabel('Unemployment Rate (%)')
ax.legend(title="Covid Period", fontsize=15)
plt.tight_layout()
plt.savefig(IMAGE_DIR_02/"malaysia_unemployment_boxplot_10years.png",dpi=300)
plt.show()



```

```python
###plt.fig(figsize=( ) is use for simple plot(no ax=ax)
### fig ,ax = plt.subplot(figsize= )  ax.set_title and axv or axhspan ,ax.set_x or yticks

# Correlation Heatmap 
plt.figure(figsize=(10, 8))
corr = df__malaysia_10years[['Unemployment Rate','GDP (billion USD)',
                            'Employment Sector: Agriculture',
                            'Employment Sector: Industry',
                            'Employment Sector: Services']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix(2012-2022)')
plt.tight_layout()
plt.savefig(IMAGE_DIR_02 / "correlation_heatmap_10years.png", dpi=300)
plt.show()
```

```python
# Sector Trends
fig, ax = plt.subplots(figsize=(12, 7))
for sector in ['Agriculture', 'Industry', 'Services']:
    sns.lineplot(data=df__malaysia_10years, x='Year', y=f'Employment Sector: {sector}', 
                 label=sector, marker='o')
ax.set_yticks(np.arange(5,70,5))    
ax.set_title('Employment Sector Trends (2012–2022)')
plt.tight_layout()
plt.savefig(IMAGE_DIR_02 / "sector_trends_10years.png", dpi=300)
```

```python
print("KEY INSIGHTS 2012-2022")

pre = df__malaysia_10years[df__malaysia_10years['covid_period'] == 0]['Unemployment Rate'].mean()
post = df__malaysia_10years[df__malaysia_10years['covid_period'] == 1]['Unemployment Rate'].mean()

print(f"• Pre-COVID average unemployment: {pre:.2f}%")
print(f"• Post-COVID average unemployment: {post:.2f}% (+{post-pre:.2f} points)")
print(f"• Highest unemployment year: {df__malaysia_10years.loc[df__malaysia_10years['Unemployment Rate'].idxmax(), 'Year']}")
```
